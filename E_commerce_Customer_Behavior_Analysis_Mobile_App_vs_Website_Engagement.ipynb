```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(df.drop(["Yearly Amount Spent","Email", "Address","Avatar"],axis=1), df["Yearly Amount Spent"], train_size=0.3,random_state=101)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred.shape
```

```python
plt.scatter(y_test,y_pred)
```

**Mean Absolute Error (MAE):**

The MAE is calculated by taking the **average of the absolute differences between the predicted and actual values.**

```python
#mean absolute error(MAE)             #when we needed absolute value
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test, y_pred)
```

```python
mae
```

**Mean Squared Error (MSE):**

The MSE is calculated by taking the **average of the squared differences between the predicted and actual values.**

```python
#mean squared error(MSE)                #when we need to compensate the +ve n -ve values from the datasets
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
mse
```

**Root Mean Squared Error (RMSE):**

The RMSE is the square root of the MSE and provides a measure of the average magnitude of the errors.



NOTE wrt to above 3 error's---These metrics are commonly used to **evaluate the accuracy of regression models**, where **lower values indicate better performance**.

```python
#Root MSE
import numpy as np
mse=mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
```

```python
rmse
```

**Residuals** 🟥

 the differences between the observed values and the predicted values in a regression analysis. You can calculate residuals by subtracting the predicted values from the actual (observed) values

 residuals will be a **NumPy array** containing the differences between the actual (y_test) and predicted (y_pred) values.

 **Positive values indicate overestimation, while negative values indicate underestimation.**

```python
#Residuals
residuals=y_test-y_pred
residuals
```

Each row in the output corresponds to a specific data point. Let's take the first row as an example:

Index 18: This is the **index or identifier of the data point** (it might represent the 18th data point in your dataset).

Residual: -3.904403: This is the difference between the actual value (y_test) and the predicted value (y_pred) for that specific data point. In this case, the actual value is approximately 3.904 less than the predicted value.

Interpretation:

If the residual is positive, it means the model underpredicted (the predicted value is lower than the actual value).

If the residual is negative, it means the model overpredicted (the predicted value is higher than the actual value).

The magnitude of the residual tells you the size of the prediction error.

For example, in the first row, the model underpredicted by approximately 3.904 units.

```python
plt.hist(residuals, bins=50, edgecolor='black')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
```
